```python
import cv2
import os

image_path = 'data/ptb-xl_images_multiprocessed/1_LVOLT.png'

if not os.path.exists(image_path):
    print(f"Error: Image file not found at '{image_path}'. Double-check the path relative to your notebook.")
else:
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print(f"Error: Could not read image from '{image_path}'. It might be corrupted or not a valid image file.")
    else:
        # Only proceed if the image was loaded successfully
        cv2.imshow('r', img)
        cv2.waitKey(0) # Wait for a key press to close the window
        cv2.destroyAllWindows() # Close all OpenCV windows
        print("Image displayed successfully.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
plt.imshow(img, cmap='gray')
```

```python
def create_thresholds(image):
    threshold_value = 127
    max_value = 255

    _, binary_image_simple = cv2.threshold(img, threshold_value, max_value, cv2.THRESH_BINARY)
    ret, binary_image_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    binary_image_adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

    return (binary_image_simple, binary_image_otsu, binary_image_adaptive)
```

```python
kernel_size = (9, 9)

gaussian_blurred = cv2.GaussianBlur(img, kernel_size, 0)
average_blurred = cv2.blur(img, kernel_size)
median_blurred = cv2.medianBlur(img, 3)

fig, axes = plt.subplots(4, 1, figsize = (15, 15))

axes[0].imshow(img, cmap = 'gray')
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(gaussian_blurred, cmap = 'gray')
axes[1].set_title('Gaussian Blurred')
axes[1].axis('off')

axes[2].imshow(average_blurred, cmap = 'gray')
axes[2].set_title('Average Blurred')
axes[2].axis('off')

axes[3].imshow(median_blurred, cmap = 'gray')
axes[3].set_title('Median Blurred')
axes[3].axis('off')
```
